# 4. Data Preparation

```python
cars_data
```

#### There is nothing to prepare because all features are Numeric.

```python
x = cars_data[["HP","VOL","SP","WT"]]
x
```

```python
y = cars_data["MPG"]
y
```

```python
# Feature Engineering Techniques can be applied later,because we need to learn it after sometime.
```

## 5. Model Building

```python
linear_model = LinearRegression()
```

# 6. Model Training

```python
linear_model.fit(x,y)
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

## 7. Model Testing

```python
y_pred = linear_model.predict(x)       # X_train(input) ah vechu y ah predict panrom
y_pred 
```

```python
# output la y_pred(MPG) vandhruku. idha training time la iruka y(MPG) kuda compare panra apo errors iruku
# y = 53.7006 , y_pred = 43.4419
```

## 8. Model Evaluation

```python
error = y - y_pred
error
```

```python
cars_data["Error"] = error     # Add a error column into the data to perform chart test
cars_data
```

#### Let's now Perform  **Test 5:** ***Homoscadacity Test***

- Homoscedasticity is a statistical assumption in regression analysis that states the variance of the **errors** (residuals-the differences between observed and predicted values) remains **constant across all levels** of the independent variables.

```python
plt.scatter( x = 'HP',y = 'Error')    #Add a error column to data (or) you have another methods too
```

```python
del cars_data['Error']
cars_data
```

```python
#Another Method

sns.scatterplot(data = cars_data, x = 'HP', y = error)  # values are acceptable in seaborn even if it is out of dataset
```

```python
sns.scatterplot(data = cars_data, x = 'VOL', y = error)
```

```python
sns.scatterplot(data = cars_data, x = 'SP', y = error)
```

```python
sns.scatterplot(data = cars_data, x = 'WT', y = error)
```

#### There is **No similar Variance** between errors - Homoscadacity Test **Failed**

#### **Test 6:** Zero Residual Mean Test

- The average of all residuals (errors) is equal to 0.
- Residual = Actual value − Predicted value

```python
sns.scatterplot(x = cars_data['MPG'], y = y_pred)
```

##### Residual is not equal to **0** - Test **Failed**
